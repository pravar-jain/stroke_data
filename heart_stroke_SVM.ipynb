```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('stroke_prediction_feature_engineered.csv')
```

```python
h_stroke = df
```

```python
df.head()
```

```python
df1 = pd.get_dummies(df,columns = ['gender','ever_married','work_type','Residence_type','smoking_status','is_diabetic','bmi_status','age_range'],drop_first = True)
```

### Scaling

```python
#seperating independent and dependent variables
x = df1.drop(['stroke'],axis = 1)
y = df1['stroke']
x.shape, y.shape
```

```python
# Importing the train test split function
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, random_state = 56)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
cols = train_x.columns
```

```python
train_x_scaled = scaler.fit_transform(train_x)
train_x_scaled = pd.DataFrame(train_x_scaled, columns=cols)
train_x_scaled.head()
```

```python
test_x_scaled = scaler.fit_transform(test_x)
test_x_scaled = pd.DataFrame(test_x_scaled, columns=cols)
test_x_scaled.head()
```

```python
train_x_scaled.shape, train_x_scaled.shape
```

---
### SVM

```python
#importing Support Vector Classifier
from sklearn.svm import SVC
```

```python
# Creating instance of SVC
svc = SVC(class_weight = 'balanced')

# Fitting the model
svc.fit(train_x_scaled, train_y)
```

```python
# Predicting over the Train Set
train_predict = svc.predict(train_x)
k = svc.score(train_x_scaled, train_y)
print('Training Score', k )
```

```python
from sklearn.metrics import classification_report
print(classification_report(train_predict, train_y))
```

```python
# Predicting over the Test Set
test_predict = svc.predict(test_x)
k = svc.score(test_x_scaled, test_y)
print('Testing Score', k )
```

```python
print(classification_report(test_predict, test_y))
```
